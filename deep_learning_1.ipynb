Individual error in the evaluation function is called as loss function. While the whole equation is termed as cost function.

We prefer MSE over MAE as loss function because it converges the gradient descent faster.

For logistic regression we use log loss. Log loss is called binary cross entropy.

We are performing mean absolute error manually.

```python
def MAE(y_actual, y_pred):
    n = len(y_pred)
    total_error = 0
    for ya, yp in zip(y_actual, y_pred):
        total_error += abs(ya - yp)
    print("Total Error:", total_error)
    mae = 1/n * total_error
    print("MAE: ", mae)
```

```python
y_pred = np.array([1,1,0,0,1])
y_actual = np.array([0.30, 0.70, 1, 0, 0.5])
MAE(y_actual, y_pred)
```

We are performing mean square error manually.

```python
def MSE(y_actual, y_pred):
    n = len(y_pred)
    total_error = 0
    for ya, yp in zip(y_actual, y_pred):
        total_error += (ya - yp)**2
    print("Total Error:", total_error)
    mse = 1/n * total_error
    print("MSE: ", mse)
```

```python
y_pred = np.array([1,1,0,0,1])
y_actual = np.array([0.30, 0.70, 1, 0, 0.5])
MSE(y_actual, y_pred)
```

```python
def log_loss(y_true, y_pred):
    epsilon = 1e-15
    y_pred_new = [max(i,epsilon) for i in y_pred]
    y_pred_new = [min(i, 1 - epsilon) for i in y_pred_new]
    y_pred_new = np.array(y_pred_new)
    return -np.mean(y_true * np.log(y_pred_new) + (1 - y_true) * np.log(1 - y_pred_new))
```

```python
def sigmoid(X):
    return 1/(1 + np.exp(-X))
```

```python
def gradient_descent(age, affordability, y_true, epochs, loss_threshold):
    w1 = w2 = 1
    bias = 0
    lr = 0.0001
    n = len(age)
    
    for i in range(epochs):
        weighted_sum = w1 * age + w2 * affordability + bias
        y_pred = sigmoid(weighted_sum)
        
        loss= log_loss(y_true, y_pred)
        
        w1d = (1/n)*np.dot(np.transpose(age), (y_pred-y_true))
        w2d = (1/n)*np.dot(np.transpose(affordability), (y_pred-y_true))
        bias_d = np.mean(y_pred-y_true)
        
        w1 = w1 - lr * w1d
        w2 = w2 - lr * w2d
        bias = bias - lr * bias_d
        
        print(f'Epoch: {i}, w1: {w1}, w2: {w2}, bias: {bias}, loss: {loss}')
        
        if loss<=loss_threshold:
            break
        
    return w1, w2, bias
```

```python
class MYNEURALNETWORK:
    def __init__(self):
        self.w1 = 1
        self.w2 = 1
        self.bias = 0
    
    def fit(self, X, y, epochs, loss_threshold):
        self.w1, self.w2, self.bias = self.gradient_descent(X['age'], X['affordibility'], y, epochs, loss_threshold)
        
    def predict(self, X_test):
        weighted_sum = self.w1 * X_test['age'] + self.w2 * X_test['affordibility'] + self.bias
        return sigmoid(weighted_sum)
    
    def gradient_descent(self, age, affordability, y_true, epochs, loss_threshold):
        w1 = w2 = 1
        bias = 0
        lr = 0.0001
        n = len(age)
    
        for i in range(epochs):
            weighted_sum = w1 * age + w2 * affordability + bias
            y_pred = sigmoid(weighted_sum)

            loss= log_loss(y_true, y_pred)

            w1d = (1/n)*np.dot(np.transpose(age), (y_pred-y_true))
            w2d = (1/n)*np.dot(np.transpose(affordability), (y_pred-y_true))
            bias_d = np.mean(y_pred-y_true)

            w1 = w1 - lr * w1d
            w2 = w2 - lr * w2d
            bias = bias - lr * bias_d
            
            if i % 50 == 0:
                print(f'Epoch: {i}, w1: {w1}, w2: {w2}, bias: {bias}, loss: {loss}')
        
            if loss<=loss_threshold:
                print(f'Epoch: {i}, w1: {w1}, w2: {w2}, bias: {bias}, loss: {loss}')
                break
        
        return w1, w2, bias
```
